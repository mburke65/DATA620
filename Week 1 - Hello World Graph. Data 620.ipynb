#### Load Packages:

```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
from pandas import DataFrame
```

#### Initialize the Graph & the Nodes:

```python
graph = nx.Graph()

nodes =  ['Andre','Beverly' , 'Carol', 'Diane', 
        'Ed', 'Fernando', 'Heather', 'Ike', 'Jane']
graph.add_nodes_from(nodes)
```

#### Add the Edge Set:

```python
edges_graph = [('Andre', 'Beverly'),
         ('Andre', 'Carol'),
         ('Andre', 'Diane'),
         ('Andre', 'Fernando'),
         ('Beverly', 'Diane'),
         ('Beverly', 'Garth'),
         ('Beverly', 'Ed'),
         ('Carol', 'Fernando'),
         ('Carol', 'Diane'),
         ('Ed', 'Diane'),
         ('Ed', 'Garth'),
         ('Diane', 'Fernando'),
         ('Diane', 'Garth'),
         ('Fernando','Garth'),
         ('Heather', 'Fernando'),
         ('Heather', 'Garth'),
         ('Heather', 'Ike'),
         ('Ike','Jane' )
        ]
graph.add_edges_from(edges_graph)
```

#### Check Connections Per Node:

```python
nx.degree(graph)
```

#### Draw Graph:

```python
#nx.draw(graph)
nx.draw(graph,node_color = 'green',with_labels = True)
plt.show()
```
